# 4. Senolíticos Inmunológicos para eliminación de LLPC  
Fundamento: Las células plasmáticas de vida larga (LLPC) IgE+ son la "fábrica" persistente de la alergia. Son resistentes a la apoptosis mediante BCL2. El uso de inhibidores como Navitoclax puede eliminar selectivamente estas células.

Ejercicio Python: Simulación de caída de población de LLPC bajo tratamiento senolítico.

```python
def simulate_llpc_decay(initial_pop, half_life_days, days, drug_efficacy=0.0):
    # La vida media de LLPC IgE+ es >49 días vs 3 días para SLPC [9]
    decay_rate = np.log(2) / half_life_days
    adjusted_rate = decay_rate + drug_efficacy # Navitoclax aumenta la tasa
    return initial_pop * np.exp(-adjusted_rate * days)

time = np.linspace(0, 365, 100)
pop_control = simulate_llpc_decay(1000, 49, time, 0.0)
pop_senolytic = simulate_llpc_decay(1000, 49, time, 0.05) # 5% incremento muerte celular

plt.plot(time, pop_control, label="Control (Persistencia)")
plt.plot(time, pop_senolytic, label="+ Navitoclax (Cura)")
plt.legend()
plt.show()
```

# 6. Machine Learning para Predicción de Respondedores (Medicina de Precisión)
 Fundamento: Un modelo de regresión logística basado en biomarcadores tempranos ($\Delta_{15w}$ de sIgG4 frente a Der p 23 y sIgE frente a Der f 1) predice la eficacia a largo plazo con un AUC de 0.896.Ejercicio Python: Implementación de la fórmula matemática del reporte.

```python
def predict_cure_probability(v1_csms, delta_sige, delta_sigg4):
    # Fórmula: y = -6.749 + 1.823(CSMS) + 0.068(ΔsIgE) + 0.003(ΔsIgG4)
    logit_y = -6.749 + (1.823 * v1_csms) + (0.068 * delta_sige) + (0.003 * delta_sigg4)
    prob = 1 / (1 + np.exp(-logit_y))
    return prob

# Paciente A: Alta respuesta inmunológica temprana
prob_a = predict_cure_probability(v1_csms=4.5, delta_sige=15.0, delta_sigg4=200.0)
print(f"Probabilidad de Tolerancia Sostenida (Paciente A): {prob_a:.2%}")
```
