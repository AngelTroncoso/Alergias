# Carga de Datos:

```python
import requests

def download_pdb_file(pdb_id, output_dir='./'):
    """
    Descarga un archivo PDB desde RCSB PDB.

    Args:
        pdb_id (str): El ID de 4 caracteres de la estructura PDB.
        output_dir (str): Directorio donde guardar el archivo descargado.
    """
    url = f"https://files.rcsb.org/download/{pdb_id}.pdb"
    filename = f"{output_dir}{pdb_id}.pdb"

    print(f"Descargando {pdb_id} de {url}...")
    response = requests.get(url, stream=True)
    response.raise_for_status() # Lanza una excepción si la solicitud no fue exitosa

    with open(filename, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"Archivo {filename} descargado exitosamente.")

# Ejemplo de uso: Descargamos el archivo 5DM2.pdb
pdb_id_to_download_list = ["5DM2", "1GP2", "1DF1", "1POA", "1K6U"]
for pdb_id in pdb_id_to_download_list:
    download_pdb_file(pdb_id)
```

## 1. Análisis de Accesibilidad y Epítopos Crípticos  
¿Cuál es el cambio en el valor de SASA para los residuos del sitio activo (Cys132, His170) tras la formación del complejo 5DM2, y qué residuos adyacentes permanecen expuestos como potenciales epítopos de unión a IgE no bloqueados por el anticuerpo actual?

```python
from Bio.PDB.SASA import ShrakeRupley
from Bio.PDB import NeighborSearch
from Bio.PDB.Polypeptide import is_aa

# --- 1. Calcular SASA para todos los residuos en la estructura ---
sr = ShrakeRupley()
sr.compute(structure, level="R") # Calcular SASA a nivel de residuo

# --- 2. Extraer SASA para los residuos del sitio activo (Cys132, His170 en la Cadena A, el TCR) ---
tcr_chain_id = 'A'
cys132_id = (' ', 132, ' ') # Formato de ID de residuo: ('hetero_flag', numero_residuo, 'codigo_insercion')
his170_id = (' ', 170, ' ')

sasa_cys132 = None
sasa_his170 = None

# Iterar a través de modelos y cadenas para encontrar la cadena TCR y los residuos objetivo
for model in structure:
    if tcr_chain_id in model:
        tcr_chain = model[tcr_chain_id]
        for residue in tcr_chain:
            if residue.get_id() == cys132_id:
                sasa_cys132 = residue.sasa
            elif residue.get_id() == his170_id:
                sasa_his170 = residue.sasa

print(f"SASA para Cys132 (Cadena {tcr_chain_id}, TCR): {sasa_cys132:.2f} Å²")
print(f"SASA para His170 (Cadena {tcr_chain_id}, TCR): {sasa_his170:.2f} Å²")

# --- 3. Identificar residuos adyacentes expuestos en el TCR (Cadena A) no bloqueados por el anticuerpo ---

# Obtener todos los átomos de todo el complejo para NeighborSearch
all_atoms = []
for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                all_atoms.append(atom)

ns = NeighborSearch(all_atoms)

# Definir umbrales de distancia y SASA
distance_threshold = 6.0 # Angstroms, define "adyacente"
sasa_exposure_threshold = 40.0 # Angstroms^2, define "expuesto" para potenciales epítopos

potential_epitope_residues = set()

# Cadenas del anticuerpo (basado en la descripción del PDB 5DM2)
antibody_chain_ids = ['H', 'L']

target_tcr_residues = []
# Asegurarse de que los residuos objetivo se encuentren antes de agregarlos a la lista
if sasa_cys132 is not None and tcr_chain_id in structure[0]:
    target_tcr_residues.append(structure[0][tcr_chain_id][cys132_id])
if sasa_his170 is not None and tcr_chain_id in structure[0]:
    target_tcr_residues.append(structure[0][tcr_chain_id][his170_id])

for target_res in target_tcr_residues:
    # Obtener todos los átomos del residuo objetivo del TCR
    target_res_atoms = [atom for atom in target_res.get_atoms()]

    for atom in target_res_atoms:
        # Encontrar todos los átomos dentro del umbral de distancia del átomo objetivo actual (puede ser de cualquier cadena)
        neighbors = ns.search(atom.get_coord(), distance_threshold)
        for neighbor_atom in neighbors:
            neighbor_residue = neighbor_atom.get_parent()
            neighbor_chain_id = neighbor_residue.get_parent().get_id()

            # Condiciones para un residuo de epítopo potencial:
            # 1. Debe ser un aminoácido
            # 2. Debe ser diferente del residuo del sitio activo objetivo
            # 3. Debe estar ubicado en el TCR (cadena A)
            # 4. Debe estar expuesto (SASA > umbral)

            if (is_aa(neighbor_residue, standard=True) and
                neighbor_residue.get_id() != target_res.get_id() and
                neighbor_chain_id == tcr_chain_id): # Asegurarse de que el vecino esté en la cadena TCR
                if hasattr(neighbor_residue, 'sasa') and neighbor_residue.sasa > sasa_exposure_threshold:
                    potential_epitope_residues.add(neighbor_residue)

print("\nResiduos adyacentes y expuestos en el TCR (Cadena A) que son potenciales epítopos no bloqueados por el anticuerpo actual:")
if potential_epitope_residues:
    # Ordenar para una salida consistente
    sorted_epitopes = sorted(list(potential_epitope_residues),
                             key=lambda r: (r.get_parent().get_id(), r.get_id()[1]))
    for res in sorted_epitopes:
        print(f"  Cadena {res.get_parent().get_id()}, Residuo {res.get_resname()}{res.get_id()[1]} (SASA: {res.sasa:.2f} Å²)")
else:
    print("  No se encontraron residuos adyacentes y expuestos en el TCR que cumplan los criterios.")
```

## 3. Simulación de Mutaciones para Hipoalérgenos
 Al realizar un escaneo de alanina in silico en los residuos de la interfaz de 5DM2, ¿qué mutaciones específicas resultan en la mayor pérdida de estabilidad del complejo alérgeno-anticuerpo sin comprometer el plegamiento global del núcleo de la proteína (medido por el mantenimiento de puentes disulfuro

```python
from Bio.PDB import NeighborSearch
from Bio.PDB.Polypeptide import is_aa
from Bio.PDB.PDBParser import PDBParser
import numpy as np

# Crear un objeto PDBParser
parser = PDBParser()

# Definir el PDB_ID para este análisis
pdb_id_for_analysis = "1GP2"
pdb_file_path = f'./{pdb_id_for_analysis}.pdb'

print(f"--- Análisis de Simulación de Mutaciones para Hipoalérgenos (Escaneo de Alanina) para {pdb_id_for_analysis} ---")

try:
    structure = parser.get_structure(pdb_id_for_analysis, pdb_file_path)
except FileNotFoundError:
    print(f"Error: El archivo {pdb_file_path} no fue encontrado. Asegúrate de que se haya descargado correctamente.")
    raise
except Exception as e:
    print(f"Error al parsear {pdb_id_for_analysis}: {e}")
    raise


# 1. Identificar residuos en la interfaz entre el TCR (Cadenas A, B) y el complejo pMHC (Cadenas C, D, E)
# Ahora ajustamos las cadenas para 1GP2
tcr_complex_chain_ids = ['A', 'B'] # Asumiendo que A y B son las cadenas del TCR en 1GP2
pmhc_complex_chain_ids = ['G'] # Asumiendo que G es la cadena principal del pMHC en 1GP2
interface_distance_cutoff = 4.0 # Angstroms, distancia típica para definir la interfaz

interface_residues_tcr = set()
interface_residues_pmhc = set()

all_atoms_tcr_complex = []
all_atoms_pmhc_complex = []

for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                if chain.get_id() in tcr_complex_chain_ids:
                    all_atoms_tcr_complex.append(atom)
                elif chain.get_id() in pmhc_complex_chain_ids:
                    all_atoms_pmhc_complex.append(atom)

# Validar que las listas de átomos no estén vacías antes de proceder
if not all_atoms_pmhc_complex:
    print(f"Error: No se encontraron átomos para las cadenas del complejo pMHC ({', '.join(pmhc_complex_chain_ids)}) en {pdb_id_for_analysis}. Verifique el contenido del archivo PDB o los IDs de las cadenas.")
    raise ValueError("No atoms found for pMHC complex chains.")
if not all_atoms_tcr_complex:
    print(f"Error: No se encontraron átomos para las cadenas del complejo TCR ({', '.join(tcr_complex_chain_ids)}) en {pdb_id_for_analysis}. Verifique el contenido del archivo PDB o los IDs de las cadenas.")
    raise ValueError("No atoms found for TCR complex chains.")

# Realizar búsqueda de vecinos para encontrar contactos
# Contactos desde TCR a pMHC
ns_pmhc = NeighborSearch(all_atoms_pmhc_complex)
for tcr_atom in all_atoms_tcr_complex:
    close_atoms = ns_pmhc.search(tcr_atom.get_coord(), interface_distance_cutoff)
    for close_atom in close_atoms:
        interface_residues_tcr.add(tcr_atom.get_parent())
        interface_residues_pmhc.add(close_atom.get_parent())

# Contactos desde pMHC a TCR (para asegurar que ambos lados de la interfaz sean capturados)
ns_tcr = NeighborSearch(all_atoms_tcr_complex)
for pmhc_atom in all_atoms_pmhc_complex:
    close_atoms = ns_tcr.search(pmhc_atom.get_coord(), interface_distance_cutoff)
    for close_atom in close_atoms:
        interface_residues_pmhc.add(pmhc_atom.get_parent())
        interface_residues_tcr.add(close_atom.get_parent())

# Filtrar por aminoácidos estándar y eliminar duplicados
interface_residues_tcr = {res for res in interface_residues_tcr if is_aa(res, standard=True)}
interface_residues_pmhc = {res for res in interface_residues_pmhc if is_aa(res, standard=True)}

print(f"\nResiduos identificados en la interfaz del TCR (Cadenas {', '.join(tcr_complex_chain_ids)}) en {pdb_id_for_analysis}:")
if interface_residues_tcr:
    sorted_tcr_interface = sorted(list(interface_residues_tcr), key=lambda r: r.get_id()[1])
    for res in sorted_tcr_interface:
        print(f"  Cadena {res.get_parent().get_id()}, Residuo {res.get_resname()}{res.get_id()[1]}")
else:
    print("  No se encontraron residuos en la interfaz del TCR.")

print(f"\nResiduos identificados en la interfaz del complejo pMHC (Cadenas {', '.join(pmhc_complex_chain_ids)}) en {pdb_id_for_analysis}:")
if interface_residues_pmhc:
    sorted_pmhc_interface = sorted(list(interface_residues_pmhc), key=lambda r: (r.get_parent().get_id(), r.get_id()[1]))
    for res in sorted_pmhc_interface:
        print(f"  Cadena {res.get_parent().get_id()}, Residuo {res.get_resname()}{res.get_id()[1]}")
else:
    print("  No se encontraron residuos en la interfaz del complejo pMHC.")

# 2. Identificar puentes disulfuro en la estructura original
def find_disulfide_bonds(structure):
    """
    Identifica puentes disulfuro buscando pares de residuos Cys que están cerca.
    Esto es una aproximación basada en distancia; para una identificación precisa,
    se deben usar los registros SSBOND del archivo PDB si están disponibles.
    """
    cysteine_sulfur_atoms = []
    for model in structure:
        for chain in model:
            for residue in chain:
                if residue.get_resname() == 'CYS' and 'SG' in residue:
                    cysteine_sulfur_atoms.append(residue['SG'])

    disulfide_bonds = []
    # Usar un umbral de distancia típico para los puentes disulfuro (~2.0-2.1 Å entre S-S)
    disulfide_distance_cutoff = 3.0 # Un poco más permisivo para la búsqueda inicial

    # Crear NeighborSearch solo con los átomos de azufre de cisteína
    if len(cysteine_sulfur_atoms) > 1:
        ns_cys_sulfur = NeighborSearch(cysteine_sulfur_atoms)
        for i, atom1 in enumerate(cysteine_sulfur_atoms):
            # Buscar vecinos para cada átomo de azufre
            neighbors = ns_cys_sulfur.search(atom1.get_coord(), disulfide_distance_cutoff)
            for atom2 in neighbors:
                # Asegurarse de no comparar el átomo consigo mismo y evitar duplicados (atom1-atom2 vs atom2-atom1)
                if (atom1.get_parent().get_full_id() != atom2.get_parent().get_full_id() and
                    (atom1.get_parent().get_full_id(), atom2.get_parent().get_full_id()) not in disulfide_bonds and
                    (atom2.get_parent().get_full_id(), atom1.get_parent().get_full_id()) not in disulfide_bonds):

                    if np.linalg.norm(atom1.get_coord() - atom2.get_coord()) < 2.5: # Distancia más estricta para SS
                        res1 = atom1.get_parent()
                        res2 = atom2.get_parent()
                        disulfide_bonds.append(tuple(sorted([
                            (res1.get_parent().get_id(), res1.get_id()),
                            (res2.get_parent().get_id(), res2.get_id())
                        ])))
    return sorted(list(set(disulfide_bonds))) # Eliminar duplicados y ordenar para consistencia

disulfide_bonds_found = find_disulfide_bonds(structure)

print("\nPuentes disulfuro identificados en la estructura original:")
if disulfide_bonds_found:
    for bond in disulfide_bonds_found:
        res1_chain, res1_id = bond[0]
        res2_chain, res2_id = bond[1]
        print(f"  Cys {res1_chain}{res1_id[1]} - Cys {res2_chain}{res2_id[1]}")
else:
    print("  No se encontraron puentes disulfuro evidentes en la estructura (buscando cisteínas cercanas).")


# 3. Explicación del Escaneo de Alanina In Silico y Evaluación de Estabilidad
print("\n--- Metodología para el Escaneo de Alanina In Silico y Evaluación de Estabilidad ---\n")
print("Para determinar qué mutaciones específicas resultan en la mayor pérdida de estabilidad del complejo TCR-pMHC (que es lo que el PDB 1GP2 representa) sin comprometer el plegamiento global del núcleo de la proteína, se seguiría el siguiente proceso, que requiere herramientas computacionales especializadas:")
print("\n**a. Escaneo de Alanina en Residuos de Interfaz:**")
print("    *   Se mutaría sistemáticamente cada residuo en las interfaces identificadas (entre TCR y pMHC) a alanina. La alanina se elige porque su cadena lateral es pequeña e inerte, eliminando interacciones específicas y minimizando cambios conformacionales.")
print("    *   Para cada mutación, se utilizarían herramientas computacionales especializadas (como **FoldX**, **Rosetta**, o software comercial como **Schrödinger Desmond/Maestro**) para predecir el cambio en la energía libre de unión (ΔΔG) del complejo.")
print("    *   Un valor positivo de ΔΔG indicaría una desestabilización de la unión del complejo (mayor pérdida de afinidad), mientras que un valor negativo indicaría una estabilización.")

print("\n**b. Evaluación del Plegamiento Global y Puentes Disulfuro:**")
print("    *   Una vez que se tengan las predicciones de ΔΔG para todas las mutaciones, se deben filtrar los resultados.")
print("    *   Estas herramientas también pueden predecir el impacto de la mutación en la estabilidad de la proteína individual (no solo en la unión del complejo).")
print("    *   Es crucial descartar mutaciones que comprometan el plegamiento global de la proteína. Esto se evaluaría monitoreando los cambios en parámetros estructurales (por ejemplo, RMSD) y, fundamentalmente, la **interrupción de puentes disulfuro**.")
print(f"    *   Para verificar el mantenimiento de los puentes disulfuro, se compararían las distancias S-S en las cisteínas de los puentes originales ({', '.join([f'Cys {b[0][0]}{b[0][1][1]}-Cys {b[1][0]}{b[1][1][1]}' for b in disulfide_bonds_found]) if disulfide_bonds_found else 'no identificados'}) en la estructura mutada respecto a la original. Las mutaciones que resulten en una ΔΔG positiva (pérdida de estabilidad del complejo) y que no afecten significativamente la estabilidad individual de las proteínas ni rompan los puentes disulfuro serían los candidatos ideales para generar hipoalérgenos.")

print("\n**c. Interpretación y Selección:**")
print("    *   Las mutaciones que muestren la mayor pérdida de estabilidad del complejo (mayor ΔΔG positiva) sin afectar la integridad de la proteína individual serán las 'mutaciones específicas' que buscas.")
print("    *   Se podría priorizar mutaciones en residuos clave que formen muchos contactos o interacciones cruciales en la interfaz, o aquellos con una alta contribución a la energía de unión.")

print("\n**En resumen:** Aunque Biopython nos permite identificar las interfaces y los puentes disulfuro, la predicción de los efectos de las mutaciones en la estabilidad requiere un software de modelado molecular más avanzado. Los resultados de este análisis computacional guiarían la selección de mutaciones para la creación de hipoalérgenos.")
```
